```python
import requests
from bs4 import BeautifulSoup
```

```python
import requests
my_url = "https://www.supremecourt.gov/oral_arguments/argument_transcript/2023"
raw_html = requests.get(my_url).content
```

```python
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(soup_doc.prettify())
```

```python
entries = soup_doc.find(class_="accordion").find_all('tr')
entries
```

```python
numbs_tag = []

for entry in entries:
    numbs = entry.find_all('a')
    if numbs:  # Check if 'a' tag exists
        numbs_tag.append(numbs)

numbs_tag
```

```python
numb_list = []
link_list = []

for numbs in numbs_tag:
    for numb in numbs:
        numb_text = numb.text.strip()
        numb_list.append(numb_text)
        link = numb['href']
        link_list.append(link)

numb_list
```

```python
name_list = []

for entry in entries:
    name = entry.find_all('span')
    if name:
        name_list.append(name[1].text)
        
name_list
```

```python
date_list = []

for entry in entries:
    date = entry.find_all('td')
    if date:
        date_list.append(date[1].text)

date_list
```

```python
entries_table = []

# for entry in entries:
#     row = {}
#     row['docket_num'] = numb_list
#     row['pdf_link'] = link_list
#     row['case_name'] = name_list
#     row['date'] = date_list
#     row=[numb_list, link_list, name_list, date_list]
#     entries_table.append(row)

# Assuming all lists are the same length
for i in range(len(numb_list)):
    row = {
        'docket_num': numb_list[i],
        'pdf_link': link_list[i],
        'case_name': name_list[i],
        'date': date_list[i]
    }
    entries_table.append(row)
    
entries_table
```

```python
import pandas as pd
df = pd.DataFrame(entries_table)
```
